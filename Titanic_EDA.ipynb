```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())

```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df = df.drop(columns=['Cabin'])
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))


numeric_df = df.select_dtypes(include=['int64','float64'])

sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="Survived", hue="Sex", data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
sns.countplot(x="Survived", hue="Pclass", data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
sns.histplot(df["Age"], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.scatterplot(x="Age", y="Fare", hue="Survived", data=df)
plt.title("Fare vs Age Colored by Survival")
plt.show()
```

EDA revealed survival was influenced by gender, class, and age.
Females and first-class passengers had the highest survival rates.
Missing values were handled using median and mode imputation,
and categorical variables were visualized for deeper insights.
