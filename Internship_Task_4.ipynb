```python
# Check missing values
print(df.isnull().sum())

# Drop or fill missing values
df = df.dropna()   # or use df.fillna(method='ffill')
```

```python

print(df.columns)



```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Age distribution vs subscription
sns.histplot(data=df, x='age', hue='y', bins=20, kde=True)
plt.title("Age vs Subscription Outcome")
plt.show()

# Job vs subscription
sns.countplot(data=df, x='job', hue='y')
plt.xticks(rotation=45)
plt.title("Job vs Subscription Outcome")
plt.show()

```

```python
# Balance vs subscription
sns.boxplot(data=df, x='y', y='balance')
plt.title("Balance vs Subscription Outcome")
plt.show()

# Housing loan vs subscription
sns.countplot(data=df, x='housing', hue='y')
plt.title("Housing Loan vs Subscription")
plt.show()
```

```python
# Duration of call vs subscription
sns.boxplot(data=df, x='y', y='duration')
plt.title("Call Duration vs Subscription Outcome")
plt.show()

# Contact type vs subscription
sns.countplot(data=df, x='contact', hue='y')
plt.title("Contact Type vs Subscription")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_df = df.select_dtypes(include=['int64','float64'])  # keep only numeric
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Numeric Features")
plt.show()

```
